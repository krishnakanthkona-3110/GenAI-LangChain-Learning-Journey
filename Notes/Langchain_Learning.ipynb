1️⃣ load_dotenv()
What it does

load_dotenv() loads environment variables from a .env file into your Python program.

You imported it from python-dotenv.

Why it is used

It allows you to store secret keys (like API keys) safely instead of writing them directly in your code.

Python automatically loads this key into the environment.

Then OpenAI libraries (like LangChain) can read it automatically.

So you don't need to write the API key in your code.

2️⃣ invoke()
What it does

invoke() sends a prompt to the LLM model and gets a response.

In your code:

result = model.invoke('write a 5 line poem on cricket')

Steps happening internally:

1️⃣ Your prompt is sent to GPT-4

2️⃣ The model processes the prompt

3️⃣ The response is returned as an AIMessage object

Returned object

invoke() does not return plain text.

It returns an object like:

AIMessage(
   content="Cricket poem text...",
   additional_kwargs={},
   response_metadata={}
)

That's why you print:

print(result.content)

to get the actual text.

In LangChain, the main methods are:

Method	Purpose

invoke()	Single input → single output

batch()	Multiple inputs

stream()	Streaming response
